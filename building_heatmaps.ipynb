```python
import pandas as pd
```

```python
data = pd.read_csv("../data/crime_processed.csv", parse_dates=["reported_date"])
data
```

Count the number of NaN values in each column

```python
location_data = data.loc[:, ("offense_category_id", "reported_date", "incident_address", "geo_x", "geo_y", "geo_lat", "geo_lon")]

location_data.astype = {
    'incident_address': str,
    'geo_x': float,
    'geo_y': float,
    'geo_lat': float,
    'geo_lon': float
}

location_data["reported_date"] = pd.to_datetime(location_data["reported_date"], format='mixed')
```

```python
na_counts = {column: sum([int(x) for x in (location_data[column].isna())]) for column in location_data}
print(na_counts)
```

```python
location_data.dropna(axis=0, how='any', inplace=True, ignore_index=True)
location_data
```

```python
na_counts = {column: sum([int(x) for x in (location_data[column].isna())]) for column in location_data}
print(na_counts)
```

```python
location_data["reported_month"] = location_data["reported_date"].apply(lambda x: x.month)
location_data["reported_year"] = location_data["reported_date"].apply(lambda x: x.year)

location_data = location_data.reindex(columns=sorted(location_data.columns))
location_data
```

```python
from statistics import mean, stdev
x = location_data["geo_x"]
y = location_data["geo_y"]
x_info = {
    'mean': mean(x),
    'std': stdev(x),
    'min': min(x),
    'max': max(x)
}

y_info = {
    'mean': mean(y),
    'std': stdev(y),
    'min': min(y),
    'max': max(y)
}

print(f"x-coordinate data =================================================\n{x_info}")
print(f"x-coordinate data =================================================\n{y_info}")
```

We need to remove all observations that are outside of denver city limits.

```python
import geopandas as gpd
from shapely.geometry.polygon import Polygon
from shapely.geometry import Point
from tqdm.auto import tqdm

denver_city_limits = gpd.read_file("../data/denver_city_limits.geojson")
denver_city_limits_polygon = Polygon(denver_city_limits.iloc[0].loc["geometry"])

tqdm.pandas(desc="Checking location data is within Denver city limits")

location_data.loc[:, ("in_city_limits")] = location_data.progress_apply(lambda x: denver_city_limits_polygon.contains(Point(x["geo_lon"], x["geo_lat"])), axis=1)

location_data
```

```python
from math import pi, log, tan
```

```python
location_data = location_data.query("in_city_limits == True")

k = 6378137
location_data["geo_x"] = location_data.loc[:, "geo_lon"].apply(lambda x: x * (k*pi/180.))
location_data["geo_y"] = location_data.loc[:, "geo_lat"].apply(lambda x: log(tan((90 + x) * pi/360.)) * k)
location_data
```

```python
from bokeh.models import ColumnDataSource, CategoricalColorMapper, Dropdown, CustomJSFilter, Column, Row
from bokeh.plotting import figure, output_notebook, show
from bokeh.palettes import TolRainbow13
import xyzservices.providers as xyz

output_notebook()

min_geo_x = min(location_data["geo_x"])
max_geo_x = max(location_data["geo_x"])
min_geo_y = min(location_data["geo_y"])
max_geo_y = max(location_data["geo_y"])

xmin = min_geo_x - 1.e-3*abs(min_geo_x)
xmax = max_geo_x + 1.e-3*abs(max_geo_x)
ymin = min_geo_y - 1.e-3*abs(min_geo_y)
ymax = max_geo_y + 1.e-3*abs(max_geo_y)

p = figure(
    x_range=(xmin, xmax),
    y_range=(ymin, ymax),
    x_axis_type="mercator",
    y_axis_type="mercator",
    align="center"
)

p.add_tile(xyz.OpenStreetMap.Mapnik)

x, y = denver_city_limits_polygon.exterior.xy


city_limits_frame = pd.DataFrame(zip(x, y), columns=["x", "y"])
city_limits_frame["x"] = city_limits_frame["x"].apply(lambda x: x * (k*pi/180.))
city_limits_frame["y"] = city_limits_frame["y"].apply(lambda x: log(tan((90 + x) * pi/360.)) * k)

city_limits_source = ColumnDataSource(city_limits_frame)
crime_points_frame = pd.DataFrame(location_data.loc[:, ("offense_category_id", "geo_x", "geo_y", "reported_month", "reported_year")]).rename({'geo_x': 'x', 'geo_y': 'y'}, axis=1)
crime_points_source = ColumnDataSource(crime_points_frame)
mapper=CategoricalColorMapper(factors=crime_points_frame["offense_category_id"].unique(), palette=TolRainbow13)

month_menu = [(str(month), str(month)) for month in sorted(crime_points_frame["reported_month"].unique())]
month_dropdown = Dropdown(label="Month", button_type="warning", menu=month_menu)

year_menu = [(str(year), str(year)) for year in sorted(crime_points_frame["reported_year"].unique())]
year_dropdown = Dropdown(label="Year", button_type="warning", menu=year_menu)

crime_type_menu = [(crime_type, crime_type) for crime_type in sorted(crime_points_frame["offense_category_id"].unique())]
crime_type_dropdown = Dropdown(label="Crime Type", button_type="warning", menu=crime_type_menu)

dropdown_row = Row(crime_type_dropdown, month_dropdown, year_dropdown, align="center")

p.line(x="x", y="y", source=city_limits_source, line_width=2, color="black", legend_label="City Limits", alpha=0.75)
p.circle(x="x", y="y", source=crime_points_source, size=0.5, color={'field': 'offense_category_id', 'transform': mapper}, legend="offense_category_id")


show(Column(dropdown_row, p))
```
